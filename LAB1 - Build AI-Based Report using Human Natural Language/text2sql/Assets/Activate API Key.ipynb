```python
import requests

API_KEY = "7n_LP1ZJkURFCQBmpHFcwCdfd0EpRjmjWRpxP8B8v_ww"
CLOUD_URL = "https://api.dataplatform.cloud.ibm.com"  # e.g., https://us-south.ml.cloud.ibm.com

# 1️⃣ Get IAM token
token_url = "https://iam.cloud.ibm.com/identity/token"
token_headers = {"Content-Type": "application/x-www-form-urlencoded"}
token_data = {
    "grant_type": "urn:ibm:params:oauth:grant-type:apikey",
    "apikey": API_KEY
}

token_response = requests.post(token_url, headers=token_headers, data=token_data)
token_response.raise_for_status()
IAM_TOKEN = token_response.json()["access_token"]

print("✅ Got IAM token")

# 2️⃣ Create task credentials
task_credentials_url = f"{CLOUD_URL}/v1/task_credentials"
task_headers = {
    "accept": "application/json",
    "Authorization": f"Bearer {IAM_TOKEN}",
    "Content-Type": "application/json"
}
task_payload = {
    "name": "SERVICE_ID_TASK_CREDENTIALS",
    "type": "iam_api_key",
    "secret": {
        "api_key": API_KEY
    }
}

task_response = requests.post(task_credentials_url, headers=task_headers, json=task_payload)
task_response.raise_for_status()

print("✅ Task credentials created:", task_response.json())
```
